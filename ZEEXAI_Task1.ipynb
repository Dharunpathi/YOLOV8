```python
import sqlite3
import cv2
import easyocr
from datetime import datetime
import os
import numpy as np
from pathlib import Path
import pandas as pd
from ultralytics import YOLO
MODEL_PATH = "/content/task1/yolo_based_numberplate/weights/best.pt"
VIDEO_IN   = "/content/Automatic Number Plate Recognition (ANPR) _ Vehicle Number Plate Recognition (1).mp4"
VIDEO_OUT  = "/content/detection_with_ocr.mp4"
IMAGES_DIR = None
PROCESS_EVERY_N = 1
CONF_THRESH = 0.25
MAX_DETECTIONS = 10
SQLITE_DB = "/content/plates.db"
model = YOLO(MODEL_PATH)
reader = easyocr.Reader(['en'])
conn = sqlite3.connect(SQLITE_DB)
c = conn.cursor()
c.execute('''
CREATE TABLE IF NOT EXISTS detections (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    plate_text TEXT,
    timestamp TEXT,
    frame_id INTEGER,
    video_time REAL,
    confidence REAL,
    x1 INTEGER, y1 INTEGER, x2 INTEGER, y2 INTEGER,
    source TEXT
)
''')
conn.commit()
print("DB ready at:", SQLITE_DB)
def ocr_plate_from_crop(crop_img, reader, preprocess=True):
    if crop_img.size == 0:
        return ""
    gray = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
    h, w = gray.shape
    scale = 2 if max(h,w) < 120 else 1
    if scale != 1:
        gray = cv2.resize(gray, (int(w*scale), int(h*scale)), interpolation=cv2.INTER_CUBIC)
    _, th = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    try:
        results = reader.readtext(th, detail=0)
    except:
        results = []
    if not results:
        try:
            results = reader.readtext(gray, detail=0)
        except:
            results = []
    combined = "".join(results).upper().replace(" ", "")
    return combined
if VIDEO_IN is not None:
    cap = cv2.VideoCapture(VIDEO_IN)
    fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
    w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out = cv2.VideoWriter(VIDEO_OUT, fourcc, fps, (w, h))
    frame_id = 0
    print("Starting video processing...")
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_id += 1
        if frame_id % PROCESS_EVERY_N != 0:
            out.write(frame)
            continue
        results = model.predict(source=[frame], conf=CONF_THRESH, iou=0.45, max_det=MAX_DETECTIONS, verbose=False)
        r = results[0]
        boxes = []
        scores = []
        if hasattr(r, "boxes") and len(r.boxes) > 0:
            try:
                boxes = r.boxes.xyxy.cpu().numpy()
                scores = r.boxes.conf.cpu().numpy()
            except:
                boxes = np.array([box.xyxy for box in r.boxes])
                scores = np.array([box.conf for box in r.boxes])
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = map(int, box[:4])
            x1, y1 = max(0, x1), max(0, y1)
            x2, y2 = min(w-1, x2), min(h-1, y2)
            conf = float(scores[i]) if i < len(scores) else None
            crop = frame[y1:y2, x1:x2]
            plate_text = ocr_plate_from_crop(crop, reader)
            label = plate_text if plate_text else f"{conf:.2f}"
            cv2.rectangle(frame, (x1,y1), (x2,y2), (0,255,0), 2)
            cv2.putText(frame, label, (x1, max(15, y1-8)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2, cv2.LINE_AA)
            timestamp = datetime.utcnow().isoformat()
            video_time = frame_id / fps
            c.execute('''
                INSERT INTO detections (plate_text, timestamp, frame_id, video_time, confidence, x1, y1, x2, y2, source)
                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
            ''', (plate_text, timestamp, frame_id, video_time, conf, x1, y1, x2, y2, VIDEO_IN))
            conn.commit()
        out.write(frame)
    cap.release()
    out.release()
    print("Done. Output video:", VIDEO_OUT)
```
